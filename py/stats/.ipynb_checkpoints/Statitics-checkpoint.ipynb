## Statistics

- Why know about statistics
- Statistics is the language of data
  - Data itself doesn’t “speak.”
  - Statistics gives analysts the tools and vocabulary to describe, summarize, and interpret what data says.
  - Terms like mean, variance, correlation, significance, and confidence are how we communicate insights objectively. 🧩 Without statistical reasoning, data analysis is just number crunching — not decision-making.
- It turns data into insight, not just reports
   - Anyone can calculate averages or draw graphs.
   - But a trained analyst can judge reliability, spot bias, and identify true trends vs. random noise.
   - Understanding sampling, variability, and error helps analysts draw valid conclusions.
   - eg: Did sales really increase because of a new campaign — or just by chance? → Only a statistical test can tell.
- It helps evaluate data quality and credibility
   - Statistics teaches how to handle: Missing or skewed data, Outliers, Sampling bias
   - This helps ensure your insights are accurate and trustworthy.
- It builds critical thinking
   - Statistics forces you to question assumptions and validate evidence.
   - You stop accepting numbers at face value and start asking:
     - “How was this data collected?”
     - “Is the difference meaningful or just random?”
- It’s essential for communication and credibility
    - Business leaders don’t just want dashboards — they want confidence in your recommendations.
    - When you explain results using statistical evidence (confidence intervals, p-values, effect sizes), your findings become defensible and credible.
- It’s the bridge between data and business
   - Statistics connects raw data with actionable insight — helping analysts:
     - Understand customers
     - Optimize operations
     - Forecast demand
     - Validate experiments
     - Support data-driven decisions
- Without Statistics : With Statistics
  - You describe data : You understand data
  - You make charts : You test hypotheses
  - You report trends : You prove insights
  - You see numbers : You see meaning


## Topics in Statistics

- Types of Data (quantitative, categorical, ordinal)
  - Population vs. Sample
  - Descriptive vs. Inferential Statistics
  - Data distributions (normal, skewed, uniform)
  - Levels of measurement (nominal, ordinal, interval, ratio)
- Descriptive Statistics & Visualization
  - Measures of Central Tendency: mean, median, mode
  - Measures of Dispersion: range, variance, standard deviation, IQR
  - Skewness and Kurtosis
  - Covariance and Correlation (Pearson, Spearman)
  - Histograms, Boxplots, Violin plots
  - Scatter plots for correlation
- Probability & Distributions
  - Basic probability rules
  - Random variables (discrete vs. continuous)
  - Probability distributions:
     - Uniform
     - Binomial
     - Poisson
     - Normal
     - Exponential
     - Sampling distributions & Central Limit Theorem
- Data Sampling & Outlier Detection
  - Sampling methods (random, stratified, systematic)
  - Outlier detection:
     - Z-score method
     - IQR method
     - Visualization-based detection (boxplot, scatter)
- Hypothesis Testing
  - Concepts
   - Null and Alternative hypotheses
   - Type I & Type II errors
   - p-values and significance level
   - One-tailed vs. Two-tailed tests
- Tests
  - One-sample t-test
  - Two-sample t-test
  - Paired t-test
  - Chi-square test (categorical data)
  - ANOVA (comparing >2 groups)
  - Non-parametric tests (Mann–Whitney U, Wilcoxon)
